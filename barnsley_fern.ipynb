### Only 10 lines of code to make a Barnsley fern !

```python
from numpy.random import uniform as u ; import matplotlib.pyplot as plt

(fig, ax), x, y, rdm = plt.subplots(1,1,figsize=(6,6)), 0, 0, u(0,1,5*10**5)
ax.set_xlim(-3,3) ; ax.set_ylim(0.4,11) ; ax.set_xticks([]) ; ax.set_yticks([])
[ax.spines[c].set_color('w') for c in ['top','bottom','left','right']]

for i in range(rdm.size):
    if rdm[i]<0.01: x, y = 0, 0.16*y
    elif rdm[i]<0.86: x, y = 0.85*x+0.04*y, -0.04*x+0.85*y+1.6
    elif rdm[i]<0.93: x, y = 0.2*x-0.26*y, 0.23*x+0.22*y+1.6
    else: x, y = -0.15*x+0.28*y, 0.26*x+0.24*y+0.44

    ax.plot(x,y,'.g',ms=0.5/y)
```

However, the code presented above is not efficient regarging its complexity. Let's vectorize it.

```python
#libraries
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1, figsize=(6, 6))

#parameters of the figure
ax.set_xlim(-3, 3) ; ax.set_ylim(0.4, 11)
ax.set_xticks([]) ; ax.set_yticks([]) #no ticks
[ax.spines[c].set_color('w') for c in ['top','bottom','left','right']] #no border color

#initialize data
N = 5*10**5
X, Y = [0], [0]
rdm = np.random.uniform(0, 1, N) #random vector of size N

# adding coordinates to X and Y
for i in range(N):
    if rdm[i] < 0.01:
        x, y = 0, 0.16 * Y[-1]
    elif rdm[i] < 0.86:
        x, y = 0.85*X[-1] + 0.04*Y[-1], -0.04*X[-1] + 0.85*Y[-1] + 1.6
    elif rdm[i] < 0.93:
        x, y = 0.2*X[-1] - 0.26*Y[-1], 0.23*X[-1] + 0.22*Y[-1] + 1.6
    else:
        x, y = -0.15*X[-1] + 0.28*Y[-1], 0.26*X[-1] + 0.24* Y[-1] +0.44
    X.append(x) ; Y.append(y)

ax.plot(X, Y, '.g', ms=0.3)
plt.show()
```
